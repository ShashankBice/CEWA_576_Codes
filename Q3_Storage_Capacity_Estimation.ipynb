```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_excel('Reservoir_Storage_Capacity.xlsx')
```

```python
df.head()
```

```python
df['S-D']=df['S']-df['D']
df['cum_S']=df.S.cumsum()
```

```python
df['cum_D']=df.D.cumsum()
df['cum_C-cum_D']=df['cum_S']-df['cum_D']
```

```python
df
```

```python
rs_max=100
df['rs_dummy']=''
df.loc[0,'rs_dummy']=rs_max+df.loc[0,'S-D']
```

```python
for i in range(1,len(df)):
    df.loc[i,'rs_dummy']=df.loc[(i-1),'rs_dummy']+df.loc[i,'S-D']
```

```python
df['rs_max']=150
```
