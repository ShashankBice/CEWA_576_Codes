```python
df['S-D']=df['S']-df['D']
df['cum_S']=df.S.cumsum()
```

```python
df['cum_D']=df.D.cumsum()
df['cum_C-cum_D']=df['cum_S']-df['cum_D']
```

```python
df['S'].sum()
```

```python
r0=0
rs_max=100
df['rs_s']=''
r=(0+df.loc[r0,'S-D'])
df.loc[0,'rs_s']=np.minimum(rs_max,r)
for i in range (1,len(df)):
    r=(df.loc[i-1,'rs_s']+df.loc[i,'S-D'])
    df.loc[i,'rs_s']=np.minimum(rs_max,r)
```
